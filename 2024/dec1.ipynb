```python
from aocd.models import Puzzle

```

```python
puzzle = Puzzle(2024, 1)
```

```python
puzzle.url
```

```python
input = puzzle.input_data
input
```

```python
row = 0
index = 0
parsedinput1 = []
parsedinput2 = []

columnbreak = input.find("   ")
parsedinput1.append(int(input[0:columnbreak]))
linebreak = input.find("\n")

while True:
    parsedinput2.append(int(input[columnbreak+3:linebreak]))
    columnbreak = input.find("   ", linebreak)
    parsedinput1.append(int(input[linebreak + 1 :columnbreak]))
    linebreak = input.find("\n", columnbreak)
    if linebreak == -1:
        break

parsedinput2.append(int(input[columnbreak + 3:]))

parsedinput1.sort()
parsedinput2.sort()

distance = 0

for index in range(len(parsedinput1)):
    delta = abs(parsedinput1[index] - parsedinput2[index])
    distance = distance + delta

print(distance)
```

```python
puzzle.answer_a = distance
```

```python
similarity = 0

for index in range(len(parsedinput1)):
    occurrence = parsedinput2.count(parsedinput1[index])
    similarity = similarity + parsedinput1[index] * occurrence

similarity
```

```python
puzzle.answer_b = similarity
```
