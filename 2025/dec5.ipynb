```python
from aocd.models import Puzzle
```

```python
puzzle = Puzzle(2025, 5)
```

```python
example_input_data_raw = puzzle.examples[0].input_data
input_data_raw = puzzle.input_data
```

```python
example_input_data_raw
```

```python
split_input = input_data_raw.split('\n\n')
fresh_ids = split_input[0].splitlines()
ingredient_ids = split_input[1].splitlines()
```

```python
all_fresh_ids = []

for item in fresh_ids:
    start = int(item.split('-')[0])
    end = int(item.split('-')[1])

    current_range = range(start, end +1 )

    all_fresh_ids.append(current_range)

```

```python
def item_in_range(item:int, range_list:list[range]) -> bool:
    for current_range in range_list:
        if item in current_range:
            return True

    return False
```

```python
count = 0

for item in ingredient_ids:
    
    if item_in_range(item=int(item), range_list=all_fresh_ids):
        count = count + 1
```

```python
count
```

```python
puzzle.answer_a = count
```

```python
def ranges_overlap(range_1:range, range_2:range) -> bool:
    
    #range 1 smaller than range 2
    if range_1[len(range_1)-1] < range_2[0]:
        return False

    #range 2 smaller than range 1
    if range_2[len(range_2)-1] < range_1[0]:
        return False

    return True
```

```python
def fix_overlap(range_1:range, range_2:range) -> range:

    range_start = min([range_1.start, range_2.start])
    range_end = max([range_1.stop, range_2.stop])

    return(range(range_start, range_end))
```

```python
#Sort list of ranges
all_fresh_ids_sorted = sorted(all_fresh_ids, key = lambda r: r.start)
```

```python
working_list = all_fresh_ids_sorted.copy()
overlap = True
index_ok = 1

while overlap:
    overlap = False

    for range_nr in range(index_ok, len(working_list)):
        range1 = working_list[range_nr -1]
        range2 = working_list[range_nr]

        if ranges_overlap(range_1=range1, range_2=range2):
            new_range = fix_overlap(range_1=range1, range_2=range2)
            
            popped = working_list.pop(range_nr-1)
            popped2 = working_list.pop(range_nr-1)
            working_list.insert(range_nr-1, new_range)
            index_ok = range_nr

            overlap = True

            break
```

```python
valid_ids_count = 0

for my_range in working_list:
    valid_ids_count = valid_ids_count + (my_range.stop - my_range.start)
```

```python
puzzle.answer_b = valid_ids_count
```
